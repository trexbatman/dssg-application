# MNIST data set analysis
Objective: Correctly classify clothes into ten categories: (0) T-shirt/top, (1) Trouser, (2) Pullover, (3) Dress, (4)
Coat, (5) Sandal, (6) Shirt, (7) Sneaker, (8) Bag, and 9) Ankle boot<br>
Details about the data set can be found on https://www.kaggle.com/zalando-research/fashionmnist

Task 1: Loading required libraries

```python
import os
import subprocess
from IPython.display import display, HTML, Image
import io
from operator import itemgetter

from TAS_Python_Utilities import data_viz
from TAS_Python_Utilities import data_viz_target
from TAS_Python_Utilities import visualize_tree

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.tree import export_graphviz
from sklearn import tree
from sklearn import metrics
from sklearn import tree
from sklearn import svm
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn import ensemble
from sklearn import linear_model
from sklearn import neighbors

%matplotlib inline
%qtconsole
```

Task 2: Loading the data set that will be used to train and validade the algorithm

```python
TrainValidationset = pd.read_csv('fashion-mnist_train.csv')
```

Task 3: Exploration of the training set

```python
TrainValidationset["label"].value_counts()
```

```python
display(TrainValidationset.select_dtypes(include=[np.number]).describe())
```

Task 4: Diving the data set into train and validate data sets (70:30)

```python
Y = TrainValidationset["label"]
```

```python
X = TrainValidationset.iloc[:,1:785]
```

```python
X_train, X_validate, y_train, y_validate \
    = train_test_split(X, Y, random_state=0, \
                                    train_size = 0.7)
```

Task 5: Building a Decision Tree Model

```python
DecisionTree1 = tree.DecisionTreeClassifier(criterion="entropy", max_depth=150)
DecisionTree1.fit(X_train,y_train)
```

Task 5.1: Assessing the performance of the DecisionTree1 on the *train set*

```python
y_pred = DecisionTree1.predict(X_train)
accuracy = \
metrics.accuracy_score(y_train, y_pred) # , normalize=True, sample_weight=None
print("Accuracy: " +  str(accuracy))
print(metrics.classification_report(y_train, y_pred))
print(metrics.confusion_matrix(y_train, y_pred))
print("Confusion Matrix")
pd.crosstab(y_train, y_pred, rownames=['True'], colnames=['Predicted'], margins=True)
```

Task 6: Evaluating the performance of the DecisionTree1 on the *validate set*

```python
y_pred = DecisionTree1.predict(X_validate)
accuracy = metrics.accuracy_score(y_validate, y_pred) 
print("Accuracy: " +  str(accuracy))
print(metrics.classification_report(y_validate, y_pred))
print(metrics.confusion_matrix(y_validate, y_pred))
print("Confusion Matrix")
pd.crosstab(y_validate, y_pred, rownames=['True'], colnames=['Predicted'], margins=True)
```

Task 7: Using Grid Search to tune the hyperparameters of DecisionTree1

```python
param_grid ={'criterion': ['gini', "entropy"], \
             'max_depth': list(range(10, 150, 20)), \
             'min_samples_split': [500] }
my_tuned_tree = GridSearchCV(tree.DecisionTreeClassifier(), \
                                param_grid, cv=2, verbose = 0, \
                            return_train_score=True)
my_tuned_tree.fit(X_train, y_train)
print("Best parameters set found on training set:")
display(my_tuned_tree.best_params_)
display(my_tuned_tree.best_score_)
display(my_tuned_tree.cv_results_)
```

Task 7.1: Evaluating DecisionTree2 (Gini and max depth 10)

```python
DecisionTree2 = tree.DecisionTreeClassifier(criterion="gini", max_depth=10)
DecisionTree2.fit(X_train,y_train)
```

Task 7.2: Assessing the performance of the DecisionTree2 on the *train set*

```python
y_pred = DecisionTree2.predict(X_train)
accuracy = \
metrics.accuracy_score(y_train, y_pred) 
print("Accuracy: " +  str(accuracy))
print(metrics.classification_report(y_train, y_pred))
print(metrics.confusion_matrix(y_train, y_pred))
print("Confusion Matrix")
pd.crosstab(y_train, y_pred, rownames=['True'], colnames=['Predicted'], margins=True)
```

Task 7.3: Assessing the performance of the DecisionTree2 on the *validate data*

```python
y_pred = DecisionTree2.predict(X_validate)
accuracy = metrics.accuracy_score(y_validate, y_pred) # , normalize=True, sample_weight=None
print("Accuracy: " +  str(accuracy))
print(metrics.classification_report(y_validate, y_pred))
print(metrics.confusion_matrix(y_validate, y_pred))
print("Confusion Matrix")
pd.crosstab(y_validate, y_pred, rownames=['True'], colnames=['Predicted'], margins=True)
```

Task 8: Experimenting with Random Forests

```python
RandomForest1 = ensemble.RandomForestClassifier(n_estimators=100, \
                                           max_features = 10,\
                                           min_samples_split=500)
RandomForest1.fit(X_train,y_train)
```

```python
y_pred = RandomForest1.predict(X_validate)
print(metrics.classification_report(y_validate, y_pred))
print("Confusion Matrix")
pd.crosstab(y_validate, y_pred, rownames=['True'], colnames=['Predicted'], margins=True)
```

Task 9: Experimenting with feature engineering 

Task 9.1: Experimenting with feature engineering (increasing number of trees)

```python
RandomForest2 = ensemble.RandomForestClassifier(n_estimators=200, \
                                           max_features = 10,\
                                           min_samples_split=500)
RandomForest2.fit(X_train,y_train)
y_pred = RandomForest2.predict(X_validate)
print(metrics.classification_report(y_validate, y_pred))
print("Confusion Matrix")
pd.crosstab(y_validate, y_pred, rownames=['True'], colnames=['Predicted'], margins=True)
```

Most of the problems are in categories (0) T-shirt/top, (2) Pullover, (4)
Coat and (6) Shirt.

Task 9.2: Experimenting with feature engineering (increasing max depth)

```python
RandomForest3 = ensemble.RandomForestClassifier(n_estimators=100, \
                                           max_features = 15,\
                                           min_samples_split=500)
RandomForest3.fit(X_train,y_train)
y_pred = RandomForest3.predict(X_validate)
print(metrics.classification_report(y_validate, y_pred))
print("Confusion Matrix")
pd.crosstab(y_validate, y_pred, rownames=['True'], colnames=['Predicted'], margins=True)
```

Task 9.3: Experimenting with feature engineering (reducing min samples split)

```python
RandomForest4 = ensemble.RandomForestClassifier(n_estimators=100, \
                                           max_features = 10,\
                                           min_samples_split=100)
RandomForest4.fit(X_train,y_train)
y_pred = RandomForest4.predict(X_validate)
print(metrics.classification_report(y_validate, y_pred))
print("Confusion Matrix")
pd.crosstab(y_validate, y_pred, rownames=['True'], colnames=['Predicted'], margins=True)
```

Task 10: Evaluating the best model (RandomForest4) on the test data set

```python
Testset = pd.read_csv('fashion-mnist_test.csv')
Y_test = Testset["label"]
X_test = Testset.iloc[:,1:785]
```

```python
y_pred = RandomForest4.predict(X_test)
print(metrics.classification_report(Y_test, y_pred))
print("Confusion Matrix")
pd.crosstab(Y_test, y_pred, rownames=['True'], colnames=['Predicted'], margins=True)
```
